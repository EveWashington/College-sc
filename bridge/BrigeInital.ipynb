# Exploratory Data Analysis about Major Completion from College Scorecard Data

Through the site 'https://collegescorecard.ed.gov/data/' one can access the Score Cards data on student completion, debt and repayment, earnings, and more, with data ranging for 1996 to 2018. The entry level analysis below focuses on the distribution of majors awarded in each graduation years across all Columbia, Barnard, and the Ivy League.

```python
def getAll(sc, deg, con):
    is_bach = (sc['HIGHDEG'] == deg)

    all_pub = sc[is_bach]

    is_public = all_pub["CONTROL"] == con

    all_pub = all_pub[is_public]

    all_pub["UGDS_Raw_BLACK"] = all_pub['UGDS'] * all_pub['UGDS_BLACK']

    all_pub[UGDS_list]

    is_black = all_pub['UGDS_Raw_BLACK'] > 100

    all_pub = all_pub[is_black]

    all_pub['blk_grad'] = all_pub['C150_4_BLACK'].fillna(all_pub['C150_4_BLACKNH'])
    all_pub['wh_grad'] = all_pub['C150_4_WHITE'].fillna(all_pub['C150_4_WHITENH'])

    all_pub['percent_gap'] = all_pub['wh_grad'] - all_pub['blk_grad']

    all_pub[['INSTNM','STABBR','percent_gap']]

    return all_pub

r2019 = getAll(sc23, 4, 1)
r2018 = getAll(sc22, 4, 1)
r2017 = getAll(sc21, 4, 1)

r2009 = getAll(sc18, 4, 1)
r2008 = getAll(sc17, 4, 1)
r2007 = getAll(sc16, 4, 1)
```

```python
def getAll1(sc, deg, con):
    is_bach = (sc['HIGHDEG'] == deg)

    all_pub = sc[is_bach]

    is_public = all_pub["CONTROL"] == con

    all_pub = all_pub[is_public]

    all_pub["UGDS_Raw_BLACK"] = all_pub['UGDS'] * all_pub['UGDS_BLACK']

    all_pub[UGDS_list]

    is_black = all_pub['UGDS_Raw_BLACK'] > 100

    all_pub = all_pub[is_black]

    all_pub['percent_gap'] = all_pub['C150_4_WHITENH'] - all_pub['C150_4_BLACKNH']

    all_pub[['INSTNM','STABBR','percent_gap']]

    return all_pub[['INSTNM','STABBR','percent_gap']]
```
