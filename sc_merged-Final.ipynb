# Part 2: 2018
With data from the Class Profile released by Columbia, detailing the incoming class (the eariliest class profile I could find was in 2013 (class of 2017) from the wayback machine website). This displays the distrubition of majors for just 2018 and compares it to the prospective majors of the class of 2018.

```python
indexed_ivy_sc22 = ivy_sc22.set_index('INSTNM')

sc22_filtered = indexed_ivy_sc22[PCIP_list].infer_objects()
sc22_filtered = sc22_filtered.rename(converter, axis=1)

sc22_filtered
```

```python
#for Columbia in 2018
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,10)
sc22_filtered.iloc[5].plot(kind='pie', ax=ax, autopct='%1.1f%%')
ax.set_ylabel('')
ax.set_xlabel('')
plt.title('College Scoreboard Breakdown of Awarded Majors for 2018')
```

```python
social_science = ['Percentage of degrees awarded in Area, Ethnic, Cultural, Gender, And Group Studies.',
                  'Percentage of degrees awarded in Philosophy And Religious Studies.',
                  'Percentage of degrees awarded in Psychology.',
                  'Percentage of degrees awarded in Social Sciences.', 
                  'Percentage of degrees awarded in History.']
engineering = ['Percentage of degrees awarded in Engineering.',
               'Percentage of degrees awarded in Computer And Information Sciences And Support Services.']
math_science = ["Percentage of degrees awarded in Natural Resources And Conservation.", 
                "Percentage of degrees awarded in Architecture And Related Services.",
                'Percentage of degrees awarded in Biological And Biomedical Sciences.',
                'Percentage of degrees awarded in Mathematics And Statistics.',
                'Percentage of degrees awarded in Physical Sciences.',
                'Percentage of degrees awarded in Health Professions And Related Programs.']
arts = ['Percentage of degrees awarded in Communication, Journalism, And Related Programs.',
        'Percentage of degrees awarded in Liberal Arts And Sciences, General Studies And Humanities.',
        'Percentage of degrees awarded in Visual And Performing Arts.',
        'Percentage of degrees awarded in Foreign Languages, Literatures, And Linguistics.',
        'Percentage of degrees awarded in English Language And Literature/Letters.',
        'Percentage of degrees awarded in Multi/Interdisciplinary Studies.']
```

```python
CU_sc22 = sc22_filtered.iloc[5]
CU_sc22 = CU_sc22.reset_index().T
CU_sc22.columns = CU_sc22.iloc[0]
CU_sc22 = CU_sc22.drop('index', axis=0)

ss_sum = 0
e_sum = 0
math_sum = 0
arts_sum = 0
for c in social_science:
    ss_sum += CU_sc22.iloc[0][c]
for c in engineering:
    e_sum += CU_sc22.iloc[0][c]
for c in math_science:
    math_sum += CU_sc22.iloc[0][c]
for c in arts:
    arts_sum += CU_sc22.iloc[0][c]

grouped = {'Social Science':ss_sum, 'Engineering':e_sum,
           'Mathematics and Natural Science':math_sum,'Arts and Humanities':arts_sum}
ivy_sc22_grouped = pd.DataFrame(grouped, index=[0])
```

```python
predicted18 = {'Social Science':22, 'Engineering':23,
           'Mathematics and Natural Science':29,'Arts and Humanities':23, 'Undecided':3}
profile_predicted18 = pd.DataFrame(predicted18, index=[0])

#PROSPECTIVE
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,10)
profile_predicted18.iloc[0].plot(kind='pie', ax=ax, autopct='%1.1f%%')
ax.set_ylabel('')
ax.set_xlabel('')
plt.title('Prospective Majors for the Class of 2018')

#ACTUAL
fig, ax = plt.subplots(1,1)
fig.set_size_inches(10,10)
ivy_sc22_grouped.iloc[0].plot(kind='pie', ax=ax, autopct='%1.1f%%')
ax.set_ylabel('')
ax.set_xlabel('')
plt.title('Actual Degrees Awarded in 2018')
```
